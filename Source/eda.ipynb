Шаг 4: Инженерия признаков (Feature Engineering)

Инженерия признаков - это процесс использования предметных знаний для выбора и преобразования наиболее значимых переменных из исходных данных при создании прогностической модели с использованием машинного обучения или статистического моделирования. Основная цель инженерии признаков - создать осмысленные данные из исходных данных.

Шаг 5: Создание признаков (Feature Creation)

Вычисление отношения активных к общему количеству аккаунтов

```python
data['Active_All'] = data['ACCOUNTS_ACTIVE'] / data['ACCOUNTS_ALL']
print(data[['ACCOUNTS_ACTIVE', 'ACCOUNTS_ALL', 'Active_All']].head())
```

Расчёт эффективности бэкапа (Backup Efficiency)

```python
data['Backup_Efficiency'] = data['Backup_size'] / data['SIZE']
print(data[[ 'SIZE', 'Backup_size', 'Backup_Efficiency']].head())
```

Расчёт kept size

```python
data['Kept_size'] = data['SIZE'] - data['Non_kept_size']
print(data[['SIZE', 'Non_kept_size', 'Kept_size']].head())
```

Шаг 7: Анализ исследовательских данных (EDA)

Описательная статистика: Рассчет основных статистических показателей (среднее, медиана, стандартное отклонение, квартили и т. д.) для каждой числовой переменной, чтобы понять распределение данных.

Визуализация данных: Использование графиков и диаграмм для визуального представления и анализа данных. Например, можно построить:

Гистограммы для числовых переменных, чтобы увидеть их распределение.

Диаграммы рассеяния для выявления корреляций между разными переменными.

Круговые диаграммы или столбчатые диаграммы для категориальных переменных, чтобы увидеть их распределение.

```python
stage_counts = data2['Stage'].value_counts()
sns.set_palette(['#b2182b','#ef8a62','#fddbc7','#f7f7f7','#d1e5f0','#67a9cf','#2166ac'])
plt.figure(figsize=(8, 6))
plt.pie(stage_counts, labels=stage_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of System Stages')
plt.axis('equal') 
plt.show()
```

```python
type_counts = data2['Type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of System Types')
plt.axis('equal')
plt.show()
```

Поиск выбросов: Проверка на наличие значений-выбросов или аномалий, которые могут искажать данные.

Корреляционный анализ: Исследование корреляции между различными переменными для определения связей между ними.

Обработка пропущенных значений: Поиск и обработка пропущенных значений в данных,чтобы неискаженно использовать их в анализе.

Группировка и сегментация: Если возможно, группировка данных по каким-либо признакам или сегментация для получения более детального представления.

Проекция будущих шагов: При анализе данных также важно предвидеть, как этот анализ может помочь в дальнейших шагах решения задачи или планирования дальнейших исследований.

Шаг 10: Преобразование данных

Применение логарифмического преобразования к переменным с высокой асимметрией

```python
data['Backup_size_log'] = np.log(data['Backup_size'] + 1)
print(data[['Backup_size', 'Backup_size_log']].head())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Transformation')
print('Skew:', round(data["Backup_size"].skew(), 2))
print('Skew:', round(data["Backup_size_log"].skew(), 2))


sns.histplot(x='Backup_size', data=data, edgecolor='black', ax=axes[0])
sns.histplot(x='Backup_size_log', data=data, edgecolor='black', ax=axes[1])

plt.tight_layout()
plt.show()
```
